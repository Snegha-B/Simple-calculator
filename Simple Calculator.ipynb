```python
# Simple Calculator 
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Select the operation you want to perform")
print("1 - Addition")
print("2 - Subtraction")
print("3 - Multiplication")
print("4 - Division")
print("5 - Modulo")

choice = input("Enter your choice: ")


if choice == "1":
    result = num1 + num2
    print("Result:", result)

elif choice == "2":
    result = num1 - num2
    print("Result:", result)

elif choice == "3":
    result = num1 * num2
    print("Result:", result)

elif choice == "4":
    if num2 != 0:
        result = num1 / num2
        print("Result:", result)
    else:
        print("Division by zero is not allowed")

elif choice == "5":
    result = num1 % num2
    print("Result:", result)

else:
    print("Invalid choice")
```
